```python
import hdr
from hdr import *
import skimage
from skimage import io
from skimage import data
from skimage.color import rgb2gray
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
imagelist = []
i_s = ["69", "70", "71", "72", "73", "74", "75", "76", "77"]

for i in i_s:
    imagelist.append(cv.imread(f"HDR inputs/living_room1/DSC_00{i}.png"))

print("Quantity of images: ", len(imagelist))
```

```python
#Plotting Images List
fig, axes = plt.subplots(4,4, figsize=(20, 20))
ax = axes.ravel()
i = 0
while i < len(imagelist):
    ax[i].imshow(imagelist[i])
    ax[i].set_title(f"image - {i}")
    i+=1

fig.tight_layout()
plt.show() 
```

```python
#Read Exposures and insert into a numpy array
txt = pd.read_csv("HDR inputs/living_room1/image_list.txt",sep="\t",header=None)
numpy = txt[1].to_numpy()
a = len(numpy)
explist = numpy[1:a].astype(float)
```

```python
#Create the log exposure array
lnexp = np.log(explist)
```

```python
#Compute HDR

output = computeHDR(imagelist, lnexp, smoothing_lambda=100., gamma=0.6)
```

```python
plt.imshow(output)
```

```python
cv.imwrite("HDR inputs/living_room1/output_hdr.png",output)
```
